```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.shape
```

```python
df.dtypes
```

```python
#receita total
round(df["Valor Venda"].sum(), 2)
```

```python
#custo total
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
#sabendo a receita e o custo, podemos calcular o lucro
df["Lucro"] = df["Valor Venda"] - df["Custo"]
df["Lucro"].head()
```

```python
round(df["Lucro"].sum(), 2)
```

```python
#criando uma coluna para o tempo de envio do produto
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
df.head()
```

### Agora queremos saber o tempo de envio para cara marca

```python
#extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.head()
```

```python
#verificando o tipo de dado da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#média do tempo de envio por marca
df.groupby("Marca")["Tempo_envio"].mean()
```

### verificando valores ausentes

```python
df.isnull().sum()
```

###lucro por ano e por marca

```python
pd.options.display.float_format = "{:20,.2f}".format
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
Lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
Lucro_ano
```

```python
#total de produtos vendidos
df.groupby(df["Produto"])["Quantidade"].sum().sort_values(ascending=False)
```

```python
#gráfico total de produtos vendidos
df.groupby(df["Produto"])["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de produtos vendidos")
plt.xlabel("Vendidos")
plt.ylabel("Produto")
```

```python
#gráfico de lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro por ano")
plt.xlabel("Ano")
plt.ylabel("Lucro")
```

```python
#selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.sample(5)
```

```python
#gráfico de lucro por mês em 2009
df_2009.groupby(df["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro por mês")
plt.xlabel("Mês")
plt.ylabel("Lucro")
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro por marcas")
plt.xlabel("Marcas")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal")
```

```python
#gráfico de lucro por classe
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro por classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal")
```
